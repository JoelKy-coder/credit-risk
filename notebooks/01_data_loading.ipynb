# Phase 2 — Data Loading

## Objective
The objective of this phase is to load the raw loan transactions and economic indicators datasets and perform a comprehensive inspection. This includes checking dataset shapes, previewing records, identifying data types, assessing memory usage, computing descriptive statistics, summarizing missing values, identifying duplicates, and analyzing the distribution of the target variable.

### Import Libraries and Load Data
We begin by importing the necessary packages and loading our datasets: `train.csv` (contains customer loan details) and `economic_indicators.csv` (contains macroeconomic features like inflation and exchange rates).

```python
# Import necessary libraries for data manipulation and system operations
import os
import pandas as pd
import sys
sys.path.append('..')
# Import custom utility function to load raw datasets
from src.utils import load_raw_data

# Define path to raw data directory
raw_dir = r'../data/raw'
# Load both training dataset (loan transactions) and economic indicators dataset
train_df, econ_df = load_raw_data(raw_dir)
```

### Dataset Shape and Preview
Let us inspect the dimensions of the datasets and preview a few rows to understand their structure.

```python
# Display the shape (dimensions) of both datasets
print(f'Train Data Shape: {train_df.shape}')
print(f'Economic Indicators Shape: {econ_df.shape}')
# Preview the first 3 rows of each dataset to understand structure and content
display(train_df.head(3))
display(econ_df.head(3))
```

### Data Types and Memory Usage
Checking data types and memory usage is critical for identifying columns that require conversion (e.g., date columns) and managing performance.

```python
# Display detailed information about training dataset (data types and memory usage)
print('--- Train Data Info ---')
train_df.info()
# Display detailed information about economic indicators dataset
print('\n--- Economic Indicators Info ---')
econ_df.info()
```

### Descriptive Statistics
We run descriptive statistics to examine the central tendency, dispersion, and range of our numerical features.

```python
# Compute and display descriptive statistics for numerical columns in training dataset
print('--- Train Data Numerical Stats ---')
display(train_df.describe())
# Compute and display descriptive statistics for numerical columns in economic indicators
print('--- Economic Indicators Numerical Stats ---')
display(econ_df.describe())
```

### Missing Values and Duplicates Summary
We inspect our datasets for missing values and duplicate rows, which will dictate our cleaning strategy.

```python
# Check for missing values in training dataset
print('Train Data Missing Values:')
print(train_df.isnull().sum())
# Count the number of duplicate rows in training dataset
print('\nTrain Data Duplicates:', train_df.duplicated().sum())
# Check for missing values in economic indicators dataset
print('\nEconomic Indicators Missing Values:')
print(econ_df.isnull().sum())
```

### Unique Value Counts
Let's check the unique value counts for key columns to understand cardinality.

```python
# Iterate over key categorical columns and display the number of unique values in each
for col in ['country_id', 'loan_type', 'New_versus_Repeat', 'lender_id']:
    print(f'Unique values in {col}: {train_df[col].nunique()}')
```

### Target Class Distribution
Understanding target class distribution is vital, especially for credit default prediction where defaults (1) are typically rare compared to repayments (0).

```python
# Count the frequency of each class (0=paid, 1=default) in the target variable
target_counts = train_df['target'].value_counts()
# Calculate the percentage distribution of target classes
target_pct = train_df['target'].value_counts(normalize=True) * 100
# Display absolute counts of each target class
print('Target Counts:')
print(target_counts)
# Display percentage distribution of target classes
print('\nTarget Percentages:')
print(target_pct)
```

### Summary of Findings
- **Training set size**: 68,654 records, 16 features.
- **Economic indicators**: 27 rows (3 countries: Ghana, Cote d'Ivoire, Kenya; 9 indicators) and years ranging from 2001 to 2023.
- **Missing values**: The training set has no missing values. The economic indicators dataset has several missing values for Kenya (specifically average precipitation and fossil fuel consumption).
- **Target imbalance**: Highly imbalanced dataset: ~98.17% paid (0) vs ~1.83% default (1). This dictates that we must use metrics like F1-score, Recall, and Precision, and use class balancing techniques during training.
